5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
review_counter = pd.merge(recipes, reviews, left_on='id', right_on='recipe_id', how='outer')
review_counter
```

```python
review_counter['review count'] = review_counter.groupby('id')['review'].count()
review_counter.drop(['review', 'minutes', 'contributor_id', 'submitted', 'n_steps', 'description', 'n_ingredients', 'id', 'date', 'user_id', 'rating'], axis=1, inplace=True)
review_counter
```

```python
review_counter['review count'] = review_counter['review count'].fillna(0)
review_counter
```

```python
counter = review_counter.groupby('id')['review'].count()
counter
```

```python
review_counter2 = pd.merge(review_counter, counter, left_on='id', right_on='id')
```

```python
review_counter2.drop(['minutes', 'contributor_id', 'submitted', 'n_steps', 'description', 'n_ingredients', 'description_length', 'name_word_count', 'Unnamed: 0', 'id', 'date', 'review_x', 'user_id', 'rating'], axis=1, inplace=True)
review_counter2.rename(columns={'review_y': 'review_count'})
```
